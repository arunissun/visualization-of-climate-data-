```python
from joblib import Parallel, delayed
import time

start = time.time()
def loop(file_number):
    return pd.read_excel(f"/Users/varungandhi/Downloads/all_years_{file_number}.xlsx")


df = Parallel(n_jobs=-1, verbose=10)(delayed(loop)(1) for file_number in range(1))
df = pd.concat(df, ignore_index=True)
end = time.time()
print("Excel//:", end - start)
```
