```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DayLocator, DateFormatter
from matplotlib.dates import HourLocator, DateFormatter
```

```python
location = '/Users/varungandhi/Desktop/data_1/'
file_1 = 'Data_ELTE_1min.pkl'
df_1min = pd.read_pickle(location + file_1)
file_2 = 'Data_BME_garden.pkl'
bme_garden = pd.read_pickle(location + file_2)
```

```python
df_1min['TIMESTAMP'] = pd.to_datetime(df_1min['TIMESTAMP'])
bme_garden['TIMESTAMP_UTC'] = pd.to_datetime(bme_garden['TIMESTAMP_UTC'])
```

```python
df_1min = df_1min.rename(columns={'TIMESTAMP': 'TIMESTAMP_UTC'})
```

```python
merge = pd.merge(bme_garden, df_1min, on='TIMESTAMP_UTC', how = 'left')
```

```python
merge['temp_Avg(1)']
```

```python
##1 and 5 m bow
es_9 = []
for i in range(len(merge.index)):
    if merge['temp_Avg(1)'][i] >= 0:
        es_9.append(6.11 * 10**((7.5 * merge['temp_Avg(1)'][i])/(237.3 + merge['temp_Avg(1)'][i])))
    else:
        es_9.append(6.11 * 10**((9.5 * merge['temp_Avg(1)'][i])/(265.5 + merge['temp_Avg(1)'][i])))
    
es_24 = [] 

for i in range(len(merge.index)):
    if merge['temp_Avg(2)'][i] >= 0:
        es_24.append(6.11 * 10**((7.5 * merge['temp_Avg(2)'][i])/(237.3 + merge['temp_Avg(2)'][i])))
    else:
        es_24.append(6.11 * 10**((9.5 * merge['temp_Avg(2)'][i])/(265.5 + merge['temp_Avg(2)'][i])))
```

```python
es_1 = []
for i in range(len(merge.index)):
    if merge['AirTC_2_Avg'][i] >= 0:
        es_1.append(6.11 * 10**((7.5 * merge['AirTC_2_Avg'][i])/(237.3 + merge['AirTC_2_Avg'][i])))
    else:
        es_1.append(6.11 * 10**((9.5 * merge['AirTC_2_Avg'][i])/(265.5 + merge['AirTC_2_Avg'][i])))
    
es_5 = [] 

for i in range(len(merge.index)):
    if merge['AirTC_1_Avg'][i] >= 0:
        es_5.append(6.11 * 10**((7.5 * merge['AirTC_1_Avg'][i])/(237.3 + merge['AirTC_1_Avg'][i])))
    else:
        es_5.append(6.11 * 10**((9.5 * merge['AirTC_1_Avg'][i])/(265.5 + merge['AirTC_1_Avg'][i])))
```

```python
merge['es_1']  = es_1
merge['es_5']  = es_5
merge['es_9']  = es_9
merge['es_24'] = es_24


merge['e_1'] = (merge['es_1'] * merge['RH_2_corr_cut'])/100
merge['e_5'] = (merge['es_5'] * merge['RH1_corr_cut'])/100

merge['e_9'] = (merge['es_9'] * merge['rh_Avg(2)'])/100
merge['e_24'] = (merge['es_24'] * merge['rh_Avg(1)'])/100
```

```python
ediff_15 = merge['e_1'] -         merge['e_5']
tdiff_15 = merge['AirTC_2_Avg'] - merge['AirTC_1_Avg']
bow_15   = 0.66 * (tdiff_15/ediff_15)

ediff_19 = merge['e_1'] -         merge['e_9']
tdiff_19 = merge['AirTC_2_Avg'] - merge['temp_Avg(1)']
bow_19   = 0.66 * (tdiff_19/ediff_19)

ediff_124 = merge['e_1'] -         merge['e_24']
tdiff_124 = merge['AirTC_2_Avg'] - merge['temp_Avg(2)']
bow_124   = 0.66 * (tdiff_124/ediff_124)


ediff_59 = merge['e_5'] -         merge['e_9']
tdiff_59 = merge['AirTC_1_Avg'] - merge['temp_Avg(1)']
bow_59   = 0.66 * (tdiff_59/ediff_59)

ediff_524 = merge['e_5'] -         merge['e_24']
tdiff_524 = merge['AirTC_1_Avg'] - merge['temp_Avg(2)']
bow_524   = 0.66 * (tdiff_524/ediff_524)

ediff_924 = merge['e_9'] -         merge['e_24']
tdiff_924 = merge['temp_Avg(1)'] - merge['temp_Avg(2)']
bow_924   = 0.66 * (tdiff_924/ediff_924)



```

```python
merge['bow_15'] = bow_15
merge['bow_19'] = bow_19
merge['bow_124'] = bow_124

merge['bow_59'] = bow_59
merge['bow_524'] = bow_524

merge['bow_924'] = bow_924

```

```python
def plotting(data_frame, col1,col2, r1,r2,col,label):
    
    
    fig,ax = plt.subplots(1,1,figsize = (20,10))
         
    ax.plot(data_frame[col1][r1:r2],data_frame[col2][r1:r2], label = label, color = col)
    ax.tick_params(axis='both', which='major', labelsize=18)
    ax.tick_params(axis='both', which='minor', labelsize=8)
    ax.tick_params(axis='x', labelrotation= 20)
        
    
    
    #axis parameters
    ax.set_ylim([-20,20])
    #ax.set_yticks(np.arange(-10,10.1,1))
    ax.legend(loc= 'upper right', prop={'size': 12}, ncol = 5)
    ax.set_ylabel(label, fontsize=20, labelpad =15 )
    ax.minorticks_on()
    ax.xaxis.set_major_formatter(DateFormatter("%d-%m-%Y %H-%M"))


plotting(merge,'TIMESTAMP_UTC', 'bow_19', r1 = 0, r2 = 5000,col = 'red',label = 'bowen ratio')
```

```python
a = merge['bow_15'].values.reshape(-1,1)
b = merge['bow_19'].values.reshape(-1,1)
c = merge['bow_124'].values.reshape(-1,1)
d = merge['bow_524'].values.reshape(-1,1)
f = np.concatenate([a,b,c,d],axis = 1)
f
```

```python
m = []
sd = []
for i in range(len(f)):
    l = np.mean(f[i])
    m.append(l)
    ll = np.std(f[i])
    sd.append(ll)

m = np.array(m)
sd = np.array(sd)
```

```python
mm = []
ss = []
for i in range(len(m)):
    if m[i] <=2 and m[i] >= -2:
        mm.append(m[i])
    else:
        mm.append(float('NAN'))
        
    if sd[i] <=2 and sd[i] >=-2:
        ss.append(sd[i])
    else:
        ss.append(float('NAN'))
```

```python
fig, ax = plt.subplots(1,1,figsize = (20,6))
ax.plot(merge['TIMESTAMP_UTC'][5000:10000],mm[5000:10000], label = 'mean' ,color = 'b')
ax.plot(merge['TIMESTAMP_UTC'][5000:10000],ss[5000:10000], label = 'sd' ,color = 'orange')
```
