```python
artists = pd.read_sql("""
            SELECT *
            FROM  artists
            """,conn)

albums = pd.read_sql("""
            SELECT *
            FROM  albums
            """,conn)

tracks = pd.read_sql("""
            SELECT *
            FROM  tracks
            """,conn)
```

```python
tracks.join(artists, on="AlbumId", how= 'left', lsuffix='t1', rsuffix='a1', sort=False)
```

```python
tracks.join(albums.set_index('AlbumId'), on="AlbumId", how= 'left', lsuffix='t1', rsuffix='a1', sort=False)


```

```python
tracks.join(albums.set_index('AlbumId'), on="AlbumId", how= 'left', lsuffix='t1', rsuffix='a1', sort=False)[['Name', 'Composer']]



```
