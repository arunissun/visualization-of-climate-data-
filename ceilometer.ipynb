```python
Jan = glob.glob(path2_jan)
Feb = glob.glob(path2_feb)
Oct = glob.glob(path2_oct)
Nov = glob.glob(path2_nov)
Dec = glob.glob(path2_dec)

def monthly_cdf(month,path, month_name, year):
    ds = xr.open_mfdataset(month[0]+ '*'+ year, combine = 'nested', concat_dim="time")
    ds.to_netcdf(path+ 'ceilo_' + month_name+ '.nc')
```

```python
monthly_cdf(Jan, path2_jan, 'jan', '2021')
monthly_cdf(Feb, path2_feb, 'feb', '2021')
monthly_cdf(Oct, path2_oct, 'oct', '2020')
monthly_cdf(Nov, path2_nov, 'nov', '2020')
monthly_cdf(Dec, path2_dec, 'dec', '2020')
```
